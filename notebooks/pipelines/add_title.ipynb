```python
from kdpipelines.Pipeline import Pipeline
from kdpipelines import Adapters, Processors, Sinks, rocksDB
from haystack.document_store.elasticsearch import ElasticsearchDocumentStore
import pandas as pd
import json
```

```python
document_store = ElasticsearchDocumentStore(host="aifb-ls3-vm1.aifb.kit.edu", username="", password="", index="document", duplicate_documents='overwrite')
```

```python
adapt = Adapters.ElasticsearchAdapter(document_store, batch_size=1000)
update_pipeline = Pipeline(adapt, batch_size=1000, cpus=10)
```

```python
#MetadataArxiveEnricher
db = rocksDB.RocksDBAdapter("arXive_metadata", "http://localhost:8089")
update_pipeline.add_processor(Processors.MetadataArxiveEnricher("arixive-id",db))
```

```python
#MetadataFieldDiscarder
fields_to_discard = ["update_date", "doi", "authors", "abstract", "versions", "comments", "report-no", "license", "id", "categories", "submitter", "journal-ref", "authors_parsed"]
update_pipeline.add_processor(Processors.MetadataFieldDiscarder(fields_to_discard))
```

```python
update_pipeline.add_sink(Sinks.ElasticsearchSink(document_store))
```

```python
print(update_pipeline)
```

```python
update_pipeline.process()
```
