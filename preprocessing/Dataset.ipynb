```python
import pandas as pd
```

```python
datasets = pd.read_json('./wikidata/datasets.json')
```

```python
datasets['id'] = datasets.apply(lambda row: row['item'].split("/")[-1], axis=1)
```

```python
datasets = datasets[~(datasets['itemLabel']==datasets['id'])]
```

```python
datasets['label_len'] = datasets.apply(lambda row: len(row['itemLabel'].split(" ")), axis=1)
```

```python
datasets = datasets[datasets['label_len']<6]
```

```python
datasets = datasets[~(datasets['instance_ofLabel'].str.contains('discography'))]
```

```python
alg_set = set(datasets['instance_ofLabel'])
d = {}
for entry in alg_set:
    d[entry] = len(datasets[datasets['instance_ofLabel'] == entry])
```

```python
d_10 = {}
for key, value in d.items():
    if value < 10:
        d_10[key] = value
```

```python
len(d_10)
```

```python
for key in d_10.keys():
    try:
        datasets = datasets[~(datasets['instance_ofLabel'].str.contains(key))]
    except:
        pass
```
